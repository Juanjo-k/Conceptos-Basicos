```python
import pandas as pd
import os
import unicodedata

def remove_accents(text):
    if isinstance(text, str):
        return ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    return text

ruta_carpeta = "/content/drive/MyDrive/CIENCIA DE DATOS- JUAN/CORTE 2/COLOMBIA/Datos"
output_path = os.path.join(ruta_carpeta, "datos_2006_2024.xlsx")

# Load dataframes
datos_2006_2018 = pd.read_excel(os.path.join(ruta_carpeta, "datos_2006_2018.xlsx"))
datos_2019_2024 = pd.read_excel(os.path.join(ruta_carpeta, "datos_2019_2024.xlsx"))

# Standardize column names (lowercase and remove leading/trailing spaces)
datos_2006_2018.columns = datos_2006_2018.columns.str.lower().str.strip()
datos_2019_2024.columns = datos_2019_2024.columns.str.lower().str.strip()

# Get column names from both dataframes
cols_2006_2018 = datos_2006_2018.columns.tolist()
cols_2019_2024 = datos_2019_2024.columns.tolist()

# Find missing columns in each dataframe and add them
missing_in_2019_2024 = list(set(cols_2006_2018) - set(cols_2019_2024))
missing_in_2006_2018 = list(set(cols_2019_2024) - set(cols_2006_2018))

for col in missing_in_2019_2024:
    datos_2019_2024[col] = None

for col in missing_in_2006_2018:
    datos_2006_2018[col] = None

# Ensure columns are in the same order before concatenating
datos_2019_2024 = datos_2019_2024[cols_2006_2018]

# Concatenate dataframes
datos_combinados = pd.concat([datos_2006_2018, datos_2019_2024], ignore_index=True)

# Convert all string columns to lowercase and remove accents
for col in datos_combinados.columns:
    if datos_combinados[col].dtype == 'object':
        try:
            datos_combinados[col] = datos_combinados[col].str.lower().apply(remove_accents)
        except AttributeError:
            # Handle cases where a column might be object but not strings (e.g., mixed types)
            pass

# Save the combined dataframe to Excel
datos_combinados.to_excel(output_path, index=False)

print(f"Proceso completado. DataFrame combinado guardado en: {output_path}")
print(f"Cantidad de filas en el DataFrame combinado: {datos_combinados.shape[0]}")
```

```python
import pandas as pd
datos = pd.read_excel("/content/drive/MyDrive/CIENCIA DE DATOS- JUAN/CORTE 2/COLOMBIA/Datos/datos_2006_2024.xlsx")
print(datos.shape)
datos.head()

# Cantidad de filas y columnas
print(f"\nCantidad de filas: {datos.shape[0]}")
print(f"Cantidad de columnas: {datos.shape[1]}")

# Cantidad de datos faltantes por columna
print("\nCantidad de datos faltantes por columna:")
print(datos.isnull().sum())

# Cantidad de cultivos únicos
cantidad_cultivos_unicos = datos['cultivo'].nunique()
print(f"\nCantidad de cultivos únicos: {cantidad_cultivos_unicos}")

# Los 10 cultivos con mayor área sembrada
import matplotlib.pyplot as plt
import seaborn as sns

area_sembrada_por_cultivo = datos.groupby('cultivo')['área sembrada (ha)'].sum().sort_values(ascending=False)
top_10_cultivos = area_sembrada_por_cultivo.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cultivos.index, y=top_10_cultivos.values, palette='viridis')
plt.title('Top 10 Cultivos con Mayor Área Sembrada')
plt.xlabel('Cultivo')
plt.ylabel('Área Sembrada (ha)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total production by department
produccion_por_departamento = datos['producción (t)'].groupby(datos['departamento']).sum().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(15, 8))
sns.barplot(x=produccion_por_departamento.index, y=produccion_por_departamento.values, palette='viridis')
plt.title('Producción Total por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Producción (toneladas)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
